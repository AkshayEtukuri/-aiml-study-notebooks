```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def df(x):              # derivative: d/dx (x²) = 2x
    return 2*x

x = np.linspace(-5, 5, 300)

plt.figure(figsize=(8, 4))
plt.plot(x, f(x), 'b-', label='f(x) = x²')
plt.plot(x, df(x), 'r--', label="f'(x) = 2x")
plt.axhline(0, color='gray', linewidth=0.5)
plt.legend(); plt.grid(True, alpha=0.3); plt.title('Derivative of x²')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 3*x

def df(x):              # d/dx (x³ - 3x) = 3x² - 3
    return 3*x**2 - 3

x = np.linspace(-3, 3, 300)

plt.figure(figsize=(8, 4))
plt.plot(x, f(x), 'b-', label='f(x) = x³ - 3x')
plt.plot(x, df(x), 'r--', label="f'(x) = 3x² - 3")
plt.axhline(0, color='gray', linewidth=0.5)

# mark where derivative = 0 (local min/max)
plt.axvline(1, color='green', linestyle=':', label='critical point x=1')
plt.axvline(-1, color='orange', linestyle=':', label='critical point x=-1')
plt.legend(); plt.grid(True, alpha=0.3); plt.title('Derivative of x³ - 3x')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)

def df(x):              # d/dx sin(x) = cos(x)
    return np.cos(x)

x = np.linspace(-2*np.pi, 2*np.pi, 300)

plt.figure(figsize=(8, 4))
plt.plot(x, f(x), 'b-', label='f(x) = sin(x)')
plt.plot(x, df(x), 'r--', label="f'(x) = cos(x)")
plt.axhline(0, color='gray', linewidth=0.5)
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],
           ['-2π', '-π', '0', 'π', '2π'])
plt.legend(); plt.grid(True, alpha=0.3); plt.title('Derivative of sin(x)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# numerical derivative using finite difference: f'(x) ≈ (f(x+h) - f(x)) / h
def f(x):
    return x**4 - 4*x**2 + x     # any function works here

def numerical_df(x, h=1e-5):
    return (f(x + h) - f(x)) / h

def analytical_df(x):             # d/dx (x⁴ - 4x² + x) = 4x³ - 8x + 1
    return 4*x**3 - 8*x + 1

x = np.linspace(-2.5, 2.5, 300)

plt.figure(figsize=(8, 4))
plt.plot(x, f(x), 'b-', label='f(x) = x⁴ - 4x² + x')
plt.plot(x, analytical_df(x), 'r--', label='analytical f\'(x)', linewidth=2)
plt.plot(x, numerical_df(x), 'g:', label='numerical f\'(x)', linewidth=2)
plt.axhline(0, color='gray', linewidth=0.5)
plt.legend(); plt.grid(True, alpha=0.3)
plt.title('Analytical vs Numerical Derivative')
plt.show()
```
