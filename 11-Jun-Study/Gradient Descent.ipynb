```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):   return x**2
def df(x):  return 2*x

x = 4.0
lr = 0.3
history = [x]

for _ in range(20):
    x = x - lr * df(x)
    history.append(x)
    if abs(df(x)) < 1e-6: break

# plot
x_range = np.linspace(-5, 5, 300)
h = np.array(history)

plt.figure(figsize=(10, 4))
plt.subplot(1,2,1)
plt.plot(x_range, f(x_range), 'b-', linewidth=2, label='f(x) = x²')
plt.plot(h, f(h), 'ro--', markersize=5, alpha=0.7, label='GD path')
plt.scatter(h[0], f(h[0]), s=100, c='green', zorder=5, label='Start')
plt.scatter(h[-1], f(h[-1]), s=100, c='red', zorder=5, label='End')
plt.legend(); plt.grid(True, alpha=0.3); plt.title('GD on f(x)=x²')

plt.subplot(1,2,2)
plt.plot(f(h), 'r-o', markersize=4)
plt.xlabel('Step'); plt.ylabel('f(x)'); plt.title('Loss over steps')
plt.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
print(f"Minimum: x={h[-1]:.6f}, f(x)={f(h[-1]):.8f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):   return x**2
def df(x):  return 2*x

x_range = np.linspace(-5, 5, 300)
configs = [
    (0.05, 'Too slow (α=0.05)', 'blue'),
    (0.4,  'Just right (α=0.4)', 'green'),
    (0.99, 'Too large (α=0.99)', 'red'),
]

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

for ax, (lr, title, color) in zip(axes, configs):
    x = 4.0
    history = [x]
    for _ in range(25):
        x = x - lr * df(x)
        history.append(x)

    h = np.array(history)
    ax.plot(x_range, f(x_range), 'b-', linewidth=2)
    ax.plot(h, f(h), color=color, marker='o', markersize=4,
            linestyle='--', alpha=0.8, label=f'{len(h)-1} steps')
    ax.scatter(h[0], f(h[0]), s=80, c='green', zorder=5)
    ax.set_title(title); ax.set_xlim(-5.5, 5.5); ax.set_ylim(-1, 20)
    ax.legend(); ax.grid(True, alpha=0.3)

plt.suptitle('Effect of Learning Rate on Gradient Descent', y=1.02)
plt.tight_layout(); plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 2D cost surface — like weights in a neural network
def f(x, y):    return x**2 + y**2
def df_dx(x):   return 2*x
def df_dy(y):   return 2*y

x, y = 3.5, -3.0     # starting point
lr = 0.2
hist_x, hist_y = [x], [y]

for _ in range(25):
    x = x - lr * df_dx(x)
    y = y - lr * df_dy(y)
    hist_x.append(x); hist_y.append(y)
    if abs(df_dx(x)) < 1e-6 and abs(df_dy(y)) < 1e-6: break

# contour plot
grid = np.linspace(-4, 4, 200)
X, Y = np.meshgrid(grid, grid)
Z = f(X, Y)

plt.figure(figsize=(7, 6))
plt.contourf(X, Y, Z, levels=30, cmap='Blues')
plt.colorbar(label='f(x,y) = x² + y²')
plt.contour(X, Y, Z, levels=15, colors='white', alpha=0.3, linewidths=0.5)
plt.plot(hist_x, hist_y, 'ro--', markersize=5, linewidth=1.5,
         alpha=0.8, label='GD path')
plt.scatter(hist_x[0], hist_y[0], s=120, c='lime',
            zorder=5, label='Start', edgecolors='white')
plt.scatter(hist_x[-1], hist_y[-1], s=120, c='red',
            zorder=5, label='End', edgecolors='white')
plt.scatter(0, 0, s=200, c='yellow', marker='*',
            zorder=5, label='True minimum (0,0)', edgecolors='black')
plt.xlabel('x'); plt.ylabel('y')
plt.title('2D Gradient Descent on f(x,y) = x² + y²')
plt.legend(); plt.show()
print(f"Final: x={hist_x[-1]:.4f}, y={hist_y[-1]:.4f}, f={f(hist_x[-1],hist_y[-1]):.6f}")
```
