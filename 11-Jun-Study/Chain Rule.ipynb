```python
import numpy as np
import matplotlib.pyplot as plt

# y = (x² + 1)³
# outer: f(u) = u³  →  f'(u) = 3u²
# inner: g(x) = x² + 1  →  g'(x) = 2x
# chain rule: dy/dx = 3(x²+1)² · 2x = 6x(x²+1)²

def y(x):
    return (x**2 + 1)**3

def dy(x):                        # chain rule result
    return 6*x * (x**2 + 1)**2

def numerical_dy(x, h=1e-5):     # verify with numerical
    return (y(x+h) - y(x)) / h

x = np.linspace(-3, 3, 300)

plt.figure(figsize=(8, 4))
plt.plot(x, y(x), 'b-', label='y = (x²+1)³')
plt.plot(x, dy(x), 'r--', label="dy/dx = 6x(x²+1)²  [chain rule]")
plt.plot(x, numerical_dy(x), 'g:', linewidth=2, label='numerical check')
plt.legend(); plt.grid(True, alpha=0.3)
plt.title('Chain Rule: y = (x² + 1)³')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# y = sin(x²)
# outer: f(u) = sin(u)  →  f'(u) = cos(u)
# inner: g(x) = x²      →  g'(x) = 2x
# chain rule: dy/dx = cos(x²) · 2x

def y(x):
    return np.sin(x**2)

def dy(x):                        # chain rule
    return np.cos(x**2) * 2*x

def numerical_dy(x, h=1e-5):
    return (y(x+h) - y(x)) / h

x = np.linspace(-3, 3, 300)

plt.figure(figsize=(8, 4))
plt.plot(x, y(x), 'b-', label='y = sin(x²)')
plt.plot(x, dy(x), 'r--', label="dy/dx = 2x·cos(x²)  [chain rule]")
plt.plot(x, numerical_dy(x), 'g:', linewidth=2, label='numerical check')
plt.legend(); plt.grid(True, alpha=0.3)
plt.title('Chain Rule: y = sin(x²)')
plt.show()

```
