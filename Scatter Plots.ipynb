```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn-v0_8-white')

data = pd.read_csv('Scatterplot_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolor = 'black', linewidth = 2, alpha = 0.50 )

plt.xscale('log')
plt.yscale('log')


plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn-v0_8-white')

data = pd.read_csv('Scatterplot_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, edgecolor = 'black', linewidth = 2, alpha = 0.50 )

plt.xscale('log')
plt.yscale('log')


plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn-v0_8-white')

data = pd.read_csv('Scatterplot_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap = 'summer', edgecolor = 'black', linewidth = 2, alpha = 0.50 )

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')


plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
