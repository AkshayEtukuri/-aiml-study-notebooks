```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm,binom

#Normal Distribution
normal_data = np.random.normal(0, 1, 1000)
plt.hist(normal_data, bins = 30 , alpha = 0.5, label = 'Normal Dist')

# Binomial Distribution 
binom_data = np.random.binomial(n=10, p=0.5,size=1000)
plt.hist(binom_data, bins =10, alpha= 0.5, label = "Binomial Dist")
plt.legend()
plt.show()
```
