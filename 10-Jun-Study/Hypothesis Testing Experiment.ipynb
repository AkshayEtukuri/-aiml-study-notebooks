```python
import numpy as np
from scipy import stats

group_a = np.random.binomial(1, 0.10, 500)
group_b = np.random.binomial(1, 0.12, 500)

t_stat, p_value = stats.ttest_ind(group_a, group_b)

if p_value < 0.05:
    print(f"Significant! P-value: {p_value:.4f}")
else:
    print(f"Not Significant.P-value: {p_value:.4f}")
```

```python
from scipy import stats

# Example B: ML Model Comparison (Accuracy on 30 test batches)
model_a = np.random.normal(0.85, 0.02, 30) # Old model
model_b = np.random.normal(0.87, 0.02, 30) # New model

t_stat, p_val = stats.ttest_ind(model_a, model_b)

print(f"P-value: {p_val:.4f}")
if p_val < 0.05:
    print("New model is significantly better!")
else:
    print("No significant improvement.")
    
```
